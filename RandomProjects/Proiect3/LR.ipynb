<h3>Pregatirea variabilelor pentru antrenare</h3>

```python
X = df['HoursSpentReading']
y = df['BooksRead']
```

```python
X = np.array(X).reshape(-1, 1)
y = np.array(y).reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=42) 
```

Aplicarea Linear Regression peste setul nostru de date

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
w = lm.coef_
b = lm.intercept_
print(f'w = {w}')
print(f'b = {b}')
#f(x) = w * x + b
```

```python
predictii = lm.predict(X_test)
```

```python
# Creează un scatter plot pentru valorile reale (X_test vs y_test)


# Plotează linia de regresie (X_test vs predicții)
plt.scatter(X_test[:, 0], y_test, color = 'blue', label = 'Date')
plt.plot(X_test, predictii, color='red', label='Regression Line')

# Adaugă etichete și titlu
plt.xlabel('Hours Spent Reading')
plt.ylabel('Books Read')
plt.title('Scatterplot of Hours Spent Reading vs Books Read')
plt.legend()

# Afișează graficul
plt.show()
```
